```python
import numpy as np
from matplotlib import pyplot as plt

def bet(n, p):
    if np.random.uniform(0,1)<p:
        return 2*n
    return 0

def simulate(initial, p, part, steps):
    n = initial
    for i in range(steps):
        d = n*part
        n = n - d
        n = n + bet(d,p)
    return n
```

```python
a = [[simulate(1, 0.8, j, 25) for i in range(100)] for j in [l/100 for l in range(100)]]
b = [[simulate(1, 0.8, j, 50) for i in range(100)] for j in [l/100 for l in range(100)]]
c = [[simulate(1, 0.8, j, 75) for i in range(100)] for j in [l/100 for l in range(100)]]
plt.plot(np.median(a,1),'blue',np.median(b,1),'green', np.median(c,1),'red')
plt.show()
```
